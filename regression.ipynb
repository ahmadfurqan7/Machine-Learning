# Regression
`Key concept in the world of machine learning`

Regression is technique that predicts the value of variable 'y' based on the values of variable 'x'. 

ex : as temperature drop sales of jacket increases

* Modelling target value based on independent prediction
* usage to determine cause-effect relationship

```python
Y = f(X) # Relantionship in regression
# x is independent variable and y is dependent variable
```

### Linear Regression
***Uses independent variable to predict the dependent variable which is continuous in the nature***

***not require usage of threshold value***

`solving cause-effect relationship problem`

Example : age, income etc

```python
y = mx + c  

# the equation for linear line where m is the slope and c is the intercept
# linear regression used to predict m and c values for dependent y variable and independent variable x
```

```python
y = (m1x1) + (m2x2) + (m3x3) + '....'+ (mnxn) + c
# plane multi independent variable
```

### Logistic Regression
***concept of probability***

***Uses independent variable to predict the dependent variable which is categorical in the nature***

***require usage of threshold value***

`solving classification problem`

Example : binary (0 or 1), Yes or No, and more

# Practical Implementation

## Linear Regression

### Import datasets

```python
# load datasets
import numpy as np
import pandas as pd
import sklearn

column_name = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
bos1 = pd.read_csv('housing.csv', delimiter=',', names=column_name) # insert column name to datasets
bos1.head(10)
```

### Cleaning Data

```python
# Clean data
# Removing NaN values
bos1.isna().sum()
```

### Splitting Model Data 

```python
# 70% for training
from sklearn.model_selection import train_test_split

X = np.array(bos1.iloc[:,0:13])
Y = np.array(bos1['MEDV'])

#testing data size is 30% of entire data
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.30, random_state=5)
```

### Using Linear Regression

```python
from sklearn.linear_model import LinearRegression
# Load first model
lr = LinearRegression()
# train the model on training data
lr.fit(x_train,y_train)
# predict the testing data 
pred_lr = lr.predict(x_test)
```

### Model Evaluation

```python
# error for linear regression
mse_lr = sklearn.metrics.mean_squared_error(y_test, pred_lr, squared=False)
print('Error for linear regression = {}'.format(mse_lr))
```

## Logistic Regression

### Import Library

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

### Import Datasets

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
```

### Look Data

```python
dataset.head()
```

### Filtering out the required columns : Age and Salary

```python
x = dataset.iloc[:,[2,3]].values
y = dataset.iloc[:,4].values
```

### Data Split for training and testing

```python
# 50% training and 50% testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=0)
```

### Feature Scalling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

### Model Fitting

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(x_train, y_train)
```

### Model Prediction

```python
y_pred = classifier.predict(x_test)
```

### Model Evaluation

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
