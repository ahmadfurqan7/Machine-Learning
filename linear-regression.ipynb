# Linear Regression
* supervised algorithm
* finding linear relationship between independent variable and dependent variable
* finds relationship between two or more continuous variables

## Type linear regression

`* Single Linear Regression`

y = m x + c

`* Multiple Linear Regression` 

y = m1 x1 + m2 x2 + ....+ c

note :

y = dependent variables

x = independent variables

## Simple Linear Regression

### Import packages and classes

```python
#import library
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([25,35,45,55,65,76])       # independent
y = np.array([5,20,14,32,22,38])        # dependent 

print(x)
print(y)
```

```python
# reshape x because this array require to be two dimensional
x = x.reshape((-1,1))
print(x)
```

### Create model and fit it

```python
model = LinearRegression()
model.fit(x,y)
```

```python
# check model fitted and interpret it
r_sq_score = model.score(x,y)
print('coefficient determination :',r_sq_score)
```

71% coefficient

```python
# intercept
print('intercept :',model.intercept_)
```

```python
# slope
print('slope :',model.coef_)
```

### Model Prediction

```python
y_pred = model.predict(x)

print('predict response :',y_pred)
```

```python
#actual values vs prediction values
print(y)
print(y_pred)
```

## Multiple Linear Regression

### Import packages and classes

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
x = [[10,21],[5,18],[15,24],[25,56],[35,11],[55,15],[55,34],[90,35]]    # multi values
y = [4,14,15,20,14,32,22,38]    # single values

print(x)
print(y)
```

in multiple linear regression, x is two dimensional and y is one dimensional array

```python
x,y = np.array(x), np.array(y)
print(x)
print(y)
```

### Create Model and fit it

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

### Get results

```python
r_square = model.score(x,y)
print('coeffficent determination :',r_square)
```

79% accuracy 

```python
print('intercept:',model.intercept_)
```

```python
print('slope:',model.coef_)
```

### Predict responses

```python
y_pred = model.predict(x)
print('Predicted response :',y_pred,sep='\n')
```

```python
# actual values vs predict values
print('Actual values :',y)
print('Predict values :',y_pred) # 79% accuracy
```
